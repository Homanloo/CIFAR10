## Section 1: Loading the Data

The required dataset is availabe both at [this link](https://www.cs.toronto.edu/~kriz/cifar.html) and in keras API. However, to save some time and avoid downloading the entire dataset everytime, I have saved them in a pickle format.

All the credit of this dataset is for: [Learning Multiple Layers of Features from Tiny Images](https://www.cs.toronto.edu/~kriz/learning-features-2009-TR.pdf), Alex Krizhevsky, 2009.

```python
import pickle as pkl

with open("..\\CIFAR10-Dataset\\train.pkl", "rb") as f:
    x_train, y_train = pkl.load(f)
    
with open("..\\CIFAR10-Dataset\\test.pkl", "rb") as f:
    x_test, y_test = pkl.load(f)
    
print(f'X train: {x_train.shape}')
print(f'Y train: {y_train.shape}')
print(f'X test: {x_test.shape}')
print(f'Y test: {y_test.shape}')

labels_name = {
    0: "airplane", 
    1: "automobile", 
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck",
}
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 5, figsize=(12, 4))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(x_train[i])
    ax.set_title(f'{labels_name[y_train[i][0]]}')
    ax.axis('off')
plt.show()
```
